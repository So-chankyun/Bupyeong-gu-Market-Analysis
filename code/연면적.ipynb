# 1. 기초구역별, 유형별 연면적 연산

```python
data = gpd.read_file('../data/전처리 파일/표제부/연면적/3. 기초구역별 건물 연면적/기초구역별_건물_연면적_5181.shp')
data.head()
```

```python
basis_area = data.loc[:,['BAS_ID','geometry']]
basis_area.drop_duplicates('BAS_ID',keep='first',inplace=True,ignore_index=True)
basis_area
```

```python
# null data 제거
data.drop(data[data['latitude'].isnull()].index,axis=0,inplace=True)
data.shape
```

```python
display(data['주용도코드'].unique())
display(len(data['BAS_ID'].unique()))
```

```python
# 사용할 row 추출
use_row = {'제1종근린생활시설','제2종근린생활시설','단독주택','공동주택'}
use_col = ['BAS_ID','대지위치','주용도코드','연면적','geometry']

use_data = data.loc[data['주용도코드'].isin(use_row),use_col].reset_index(drop=True)
use_data['주용도코드'].unique()
```

# 2. 사용하지 않는 Row들의 데이터

```python
not_use_data = data.loc[~data['주용도코드'].isin(use_row), use_col]
not_use_data
```

```python
not_use = not_use_data.groupby(['BAS_ID','주용도코드'])['연면적'].sum().to_frame().reset_index()
not_use
```

# 3. 사용 데이터

```python
# 근린생활시설 용어 통합
use_row_index = use_data[use_data['주용도코드'].str.contains('종')].index

for idx, row in use_data.iterrows():
    if idx in use_row_index:
        use_data.iloc[idx,2] = '근린생활시설'
```

```python
use_data['주용도코드'].unique()
```

```python
# 피벗테이블 생성
agg_data = pd.pivot_table(use_data, values = '연면적',columns='주용도코드', index='BAS_ID', aggfunc=np.sum
                        ,fill_value=0)
agg_data.reset_index(inplace=True)
agg_data
```

# 4. 총연면적 생성

```python
total_building_area = data.loc[:,use_col].groupby('BAS_ID')['연면적'].sum().to_frame().reset_index()
total_building_area
```

```python
# 기초구역별 총연면적과 공동, 근린, 단독 주택의 연면적을 병합
add_total_building_area = agg_data.merge(total_building_area,how='right',right_on='BAS_ID',left_on='BAS_ID').fillna(0)

# 공동,근린,단독 주택의 비율 연산
add_total_building_area.rename(columns={'연면적':'총연면적'},inplace=True)
add_total_building_area['공동주택 비율'] = (add_total_building_area['공동주택'] / add_total_building_area['총연면적']).apply(lambda x : round(x,3)*100)
add_total_building_area['근린생활시설 비율'] = (add_total_building_area['근린생활시설'] / add_total_building_area['총연면적']).apply(lambda x : round(x,3)*100)
add_total_building_area['단독주택 비율'] = (add_total_building_area['단독주택'] / add_total_building_area['총연면적']).apply(lambda x : round(x,3)*100)

# 널값 채우고 출력
add_total_building_area
```

```python
result = basis_area.merge(add_total_building_area,how='left',right_on='BAS_ID',left_on='BAS_ID').fillna(0)
result.sort_values(by='BAS_ID',inplace=True)
result.reset_index(inplace=True,drop=True)
# result= result.fillna(0)
result
```

```python
result.to_csv('../data/전처리 파일/표제부/연면적/기초구역별_건물_연면적(집계후,1126).csv',index=False)
```

```python
cwd = os.getcwd()
file_path = os.path.join(cwd,'data/건물 리스트/201912_건물DB_전체분/build_incheon.txt')

f = open(file_path)
df = pd.DataFrame()
i = 0

with open(file_path) as f:
    for line in f.readlines():
        # 라인하나 불러옴
        line = line.replace("|",",")
    
        # 콤마를 기준으로 데이터를 나눔
        val_split = line.split(sep=',')
    
        if val_split[2] != '부평구':
            continue
    
        df_tmp = pd.DataFrame.from_dict([{'시도명':val_split[1],
                              '시군구명':val_split[2],
                              '법정읍면동명':val_split[3],
                              '법정리명':val_split[4],
                              '본번':val_split[6],
                              '부번':val_split[7],
                              '기초구역번호':val_split[27]}])
        
    
        df = pd.concat([df,df_tmp],axis=0,ignore_index=True)
        i += 1
        
        if i == 1000:
            display(val_split[2])
            display(i)

df.head()
```

## 개별공시지가 전처리

```python
file_path = os.path.join(cwd,'data/전처리 파일/개별공시지가(수정본).csv')
df = pd.read_csv(file_path)
df.head()
```

```python
change_row_idx = df[df['기초구역번호'].isin({'0','1','오피스텔','한국아파트'})].index[1:]

# 3258, 21404
# 17700, 21313
# 18483,21318
# 18968, 21320
# 18970, 21320
# 19904, 21333
```

```python
# 옳지 않은 기초구역번호 수정
basis_area_dict = {'3258':'21404','17700':'21313','18483':'21318','18968':'21320','18970':'21320'
                  ,'19904':'21333'}

for idx in change_row_idx:
    df.iloc[idx,5] = basis_area_dict[str(idx)]
    
df = df.drop(1542,axis=0)
df = df.reset_index(drop=True)
```

```python
basis_area = np.sort(df['기초구역번호'].unique())
basis_area
```

```python
df.isnull().sum()
```

```python
# 필요없는 컬럼제거
price_data = df
# df.drop('법정리명',axis=1)
```

```python
price_data.info()
```

```python
table = str.maketrans(',원','  ')

def remove_char(price):
    result = price.translate(table).replace(' ','')
    return result

for col in range(6,9):
    price_data.iloc[:,col] = price_data.iloc[:,col].astype(str)
    price_data.iloc[:,col] = price_data.iloc[:,col].apply(lambda x : remove_char(x)).astype(int)

price_data.head()
```

```python
price_data.info()
```

```python
pd.set_option("max_r", 20)
```

```python
price_data['차분평균']= [round(np.diff(list(row[6:9])).mean(),3) for idx, row in price_data.iterrows()]
price_data['차분평균']
```

```python
result = StandardScaler().fit_transform(price_data['차분평균'].values.reshape(-1,1))
price_data['차분평균(정규화)'] = result
```

```python
sns.histplot(price_data,x='차분평균(정규화)',kde=True,color='violet')
plt.show()
```

```python
pd.set_option("max_rows",100)
```

```python
price_data[price_data['차분평균'] == 0]
```

```python
price_data.columns
```

```python
all_zero = []

for idx, row in price_data.iterrows():
    year_17 = row[6]
    year_18 = row[7]
    year_19 = row[8]
    
    if (year_17+year_18+year_19) == 0:
        all_zero.append(idx)

display(price_data.iloc[all_zero])
display(len(all_zero))
```

```python
price_remove_zero = price_data.drop(all_zero,axis=0).reset_index(drop=True)
price_remove_zero.isnull().sum()
```

```python
result = price_remove_zero.groupby('기초구역번호')['차분평균'].mean().to_frame().reset_index()
result.head()
```

## 기초구역정보 로드

```python
basis_area = gpd.read_file('./data/전처리 파일/부평구_기초구역_5181/부평구_기초구역_5181.shp')
basis_area
```

# 2. 상대적 상권 유형

```python
open_kind_data = open_merge_data.copy()
close_kind_data = close_merge_data.copy()
```

```python
# 상대적인 상권활성도 컬럼-open
for col in list(open_kind_data)[:16]:
    for idx, data in enumerate(open_kind_data[col]):
        mean = open_kind_data[col].mean()
        # 평균이 0보다 큰 경우
        if mean > 0:
            if data > mean:
                open_kind_data.loc[idx,'open_kind_'+col] = 'H_p'
            else:
                open_kind_data.loc[idx,'open_kind_'+col] = 'L_p'
        # 평균이 0보다 같거나 작은 경우
        else:
            if data > mean:
                open_kind_data.loc[idx,'open_kind_'+col] = 'L_m'
            else:
                open_kind_data.loc[idx,'open_kind_'+col] = 'H_m'
```

```python
open_kind_data = open_kind_data.loc[:,'BAS_ID':]
```

```python
# 상대적인 상권활성도 컬럼-open
for col in list(close_kind_data)[:16]:
    for idx, data in enumerate(close_kind_data[col]):
        mean = close_kind_data[col].mean()

        # 평균이 0보다 큰 경우
        if mean > 0:
            if data > mean:
                close_kind_data.loc[idx,'close_kind_'+col] = 'H_p'
            else:
                close_kind_data.loc[idx,'close_kind_'+col] = 'L_p'
        # 평균이 0보다 같거나 작은 경우
        else:
            if data > mean:
                close_kind_data.loc[idx,'close_kind_'+col] = 'L_m'
            else:
                close_kind_data.loc[idx,'close_kind_'+col] = 'H_m'
```

```python
close_kind_data = close_kind_data.loc[:,'BAS_ID':]
```

```python
int_kind_data = open_kind_data.merge(close_kind_data,how='left',right_on='BAS_ID',left_on='BAS_ID')

for col in list(int_kind_data.iloc[:,2:18]):
    split_col = col.split('_')
    if len(split_col) == 3:
        int_kind_data[split_col[-1]] = int_kind_data[col]+int_kind_data['_'.join(['close',split_col[1],split_col[2]])]
    elif len(split_col) == 4:
        int_kind_data['_'.join([split_col[2],split_col[3]])] = int_kind_data[col]+int_kind_data['_'.join(['close',split_col[1],split_col[2]
                                                                                        ,split_col[3]])]
```

```python
al_LL = ['H_pH_p','L_mH_p','H_pL_m','L_mL_m']
al_LH = ['H_pL_p','H_pH_m','L_mL_p','L_mH_m']
al_HL = ['L_pH_p','L_pL_m','H_mH_p','H_mL_m']
al_HH = ['L_pL_p','L_pH_m','H_mL_p','H_mH_m']

# al_kind = [(al_LL,'창업이 더 저조, 폐업이 더 저조'),
#            (al_LH,'창업이 더 저조, 폐업이 덜 저조'),
#            (al_HL,'창업이 덜 저조, 폐업이 더 저조'),
#            (al_HH,'창업이 덜 저조, 폐업이 덜 저조')]

# al_kind = [(al_LL,'상대적 상권 순환 저조'),
#            (al_LH,'상대적 상권 축소'),
#            (al_HL,'상대적 상권 확장'),
#            (al_HH,'상대적 상권 순환 활발')]

al_kind = [(al_HH,'1'),
           (al_LL,'2'),
           (al_LH,'3'),
           (al_HL,'4')
           ]

def classify_kind(kind):
    for kind_set, msg in al_kind:
        if kind is np.nan:
            return ''

        if kind in kind_set:
            return msg
```

```python
interpret_kind = pd.DataFrame(columns=int_kind_data.columns,index=int_kind_data.index)

for i, row in int_kind_data.iloc[:,3:].iterrows():
    for j, kind in enumerate(row):
        interpret_kind.iloc[i,j+3] = classify_kind(kind)

interpret_kind['BAS_ID'] = int_kind_data['BAS_ID']
interpret_kind['open_15cluster'] = int_kind_data['open_15cluster']
interpret_kind['close_15cluster'] = int_kind_data['close_15cluster']
```

```python
open_group = pd.DataFrame(index=sorted(list(int_kind_data['open_15cluster'].unique())))
close_group = pd.DataFrame(index=sorted(list(int_kind_data['close_15cluster'].unique())))
display(open_group)
display(close_group)
```

```python
table = str.maketrans('{}\'\'','    ')

for col in list(interpret_kind)[3:]:
    open_msg_group = interpret_kind.groupby('open_15cluster')[col].apply(lambda x:"%s" % '|'.join(x))
    close_msg_group = interpret_kind.groupby('close_15cluster')[col].apply(lambda x:"%s" % '|'.join(x))
    open_msg = []
    close_msg = []

    for i in open_msg_group.index:
        open_msg.append(set(open_msg_group.loc[i].split('|')))
    open_group.loc[:,col] = pd.Series([str(msg).translate(table).replace(' ','') for msg in open_msg])
    open_group.loc[:,col]=open_group.loc[:,col].apply(lambda x: x[1:] if x.startswith(',') else x)

    for i in close_msg_group.index:
        close_msg.append(set(close_msg_group.loc[i].split('|')))

    close_group.loc[:,col] = pd.Series([str(msg).translate(table).replace(' ','') for msg in close_msg],index=close_msg_group.index)
    close_group.loc[:,col]=close_group.loc[:,col].apply(lambda x: x[1:] if x.startswith(',') else x)
```

# 4. 군집별 유사도 - 평균 영업 기간

```python
sign_to_binary = []
nrows, ncols = open_result.shape

for i in range(nrows):
    str_bi = ''
    for j in range(ncols):
        if open_result.iloc[i,j] > 0:
            str_bi += '1'
        else:
            str_bi += '0'
    sign_to_binary.append(str_bi)
```

```python
sign_to_binary
```

```python
binary_matrix = pd.DataFrame(columns=open_result.index, index=open_result.index)

for i in range(nrows):
    for j in range(nrows):
        if sign_to_binary[i] == sign_to_binary[j]:
            binary_matrix.iloc[i,j] = 1
        else:
            binary_matrix.iloc[i,j] = 0
```

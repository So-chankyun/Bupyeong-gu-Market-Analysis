```python
preprocesssing_path = '../data/전처리 파일/아파트 가구수'
apart_data = gpd.read_file(os.pathd.join(preprocesssing_path,'기초구역별_아파트가구수_5181_1126.shp'))
apart_data
```

```python
null_data = apart_data.drop_duplicates(keep='first')
null_idx = null_data[null_data['latitude'].isnull()].index
```

```python
null_data = null_data.drop(null_idx,axis=0)
```

```python
basis_data = apart_data[['BAS_ID','geometry']].drop_duplicates('BAS_ID',keep='first',ignore_index=True)
basis_data
```

```python
use_col = ['BAS_ID','세대수']
use_data = null_data.loc[:,use_col]
use_data
```

```python
use_data.dropna(inplace=True)
```

```python
result = use_data.groupby('BAS_ID')['세대수'].sum().to_frame().reset_index()
result
```

```python
loss_bas_id = set(basis_data['BAS_ID']).difference(set(result['BAS_ID']))


for bas_id in list(loss_bas_id):
    new_data = pd.DataFrame.from_dict([{'BAS_ID':bas_id,'세대수':0}])
    result = pd.concat([result,new_data],axis=0)

add_null_data_result = result.sort_values(by='BAS_ID')
add_null_data_result
```

```python
merge_data = basis_data.merge(add_null_data_result,how='inner',right_on='BAS_ID',left_on='BAS_ID')
merge_data.sort_values(by='BAS_ID',inplace=True)
```

```python
merge_data.to_csv(os.path.join(preprocesssing_path,'기초구역별_아파트가구수(집계후)_5181_1129.csv')
                            ,index=False)
```
